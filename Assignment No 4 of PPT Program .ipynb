Q.1 Given three integer arrays arr1, arr2 and arr3 sorted in strictly increasing order, return a sorted array of only the integers that
appeared in all three arrays.<br>
Example 1:<br>
Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]<br>
Output: [1,5]<br>
Explanation: Only 1 and 5 appeared in the three arrays.

```python
# Ans:
def common_elements(arr1,arr2,arr3):
    s=[]
    for i in arr1:
        if i in arr2:
            if i in arr3:
                s.append(i)
    return s
arr1=[1,2,3,4,5]
arr2 = [1,2,5,7,9]
arr3 = [1,3,4,5,8]
output=common_elements(arr1,arr2,arr3)
print("Output:",output)
```

Q.2 Given two 0-indexed integer arrays nums1 and nums2, return a list answer of size 2 where:<br>
• answer[0] is a list of all distinct integers in nums1 which are not present in nums2. <br>
• answer[1] is a list of all distinct integers in nums2 which are not present in nums1.<br>

Note that the integers in the lists may be returned in any order.<br>
Example 1:<br>
Input: nums1 = [1,2,3], nums2 = [2,4,6]<br>

Output: [[1,3],[4,6]]<br>

Explanation:<br>
For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas<br>
nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore,<br>
answer[0] = [1,3].<br>

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas<br>
nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore,<br>
answer[1] = [4,6]

```python
def numsNotPresent(nums1,nums2):
    s=[]
    s1=[]
    s2=[]
    for i in nums1:
        if i not in nums2:
            s1.append(i)
   
    for j in nums2:
        if j not in nums1:
            s2.append(j)
    s=[s1]+[s2]
    return s

nums1 = [1,2,3]
nums2 = [2,4,6]
output=numsNotPresent(nums1,nums2)
print("Output:",output)
```

Q.4 Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi)<br>
for all i is maximized. Return the maximized sum.<br>
Example 1:<br>
Input: nums = [1,4,3,2]<br>
Output: 4<br>
Explanation: All possible pairings (ignoring the ordering of elements)
are:<br>
1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3<br>
2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3<br>
3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4<br>
So the maximum possible sum is 4

```python
# Ans:
def maxSum(nums):
    nums.sort()
    max_sum=0
    for i in range(0,len(nums),2):
        max_sum+=nums[i]
    
    return max_sum

nums=[1,4,3,2]
max_sum=maxSum(nums)
print("Maximum Sum:",max_sum)
```

Q.5 You have n coins and you want to build a staircase with these coins.<br>
The staircase consists of k rows where the ith row has exactly i coins.<br>
The last row of the staircase may be incomplete.<br>
Given the integer n, return the number of complete rows of the staircase you will build.<br>
Example 1:<br>
""<br>
Input: n = 5<br>
Output: 2<br>
Explanation: Because the 3rd row is incomplete, we return 2.

```python
def arrangCoins(n):
    k=1
    while (k*(k+1))//2<=n:
        k+=1
    return k-1
n=5
s=arrangCoins(n)    
print("Output:",s)
```

Q.6 Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.<br>
Example 1:<br>
Input: nums = [-4,-1,0,3,10]<br>
Output: [0,1,9,16,100]<br>
Explanation: After squaring, the array becomes [16,1,0,9,100].<br>
After sorting, it becomes [0,1,9,16,100]

```python
# Ans:
def squredArray(nums):
    l=[]
    for i in nums:
        l.append(i*i)
    l.sort()
    return l

nums = [-4,-1,0,3,10]
o=squredArray(nums)
print("Output:",o)       
```

Q.7 You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be<br>
incremented by one for all 0 <= x < ai and 0 <= y < bi.<br>
Count and return the number of maximum integers in the matrix after performing all the operations<br>
Example 1:<br>
<br>
Input: m = 3, n = 3, ops = [[2,2],[3,3]]<br>
Output: 4<br>
Explanation: The maximum integer in M is 2, and there are four of it
in M. So return 4.

```python
# Ans:
def maxCount(m,n,ops):
    min_ai=m
    min_bi=n
    for i in ops:
        min_ai=min(min_ai,i[0])
        min_bi=min(min_bi,i[1])
        
    return min_ai*min_bi

m=3
n=3
ops=[[2,2],[3,3]]
max_integers= maxCount(m,n,ops)
print("Output:",max_integers)
```

Q.8 Given the array nums consisting of 2n elements in the form<br>
[x1,x2,...,xn,y1,y2,...,yn].<br>
Return the array in the form [x1,y1,x2,y2,...,xn,yn].<br>
Example 1:<br>
Input: nums = [2,5,1,3,4,7], n = 3<br>
Output: [2,3,5,4,1,7]<br>
Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the
answer is [2,3,5,4,1,7]

```python
def shuffle(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])
        result.append(nums[i + n])
    return result

nums = [2, 5, 1, 3, 4, 7]
n = 3
result = shuffle(nums, n)
print("Output:", result)
```
